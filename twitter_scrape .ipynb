```python
#searching tweets from my timeline 
tweets_list = []
public_tweets = api.home_timeline(1000)
for tweet in public_tweets:
    tweets_list.append(tweet._json)
#     print({tweet.text:tweet.user.screen_name})

len(tweets_list)
```

```python
#write list of tweets into a txt file 
with open('tweet_json_Barca.txt', 'w') as file:
        file.write(json.dumps(tweets_list, indent=4))
```

```python
#create a dataframe from json tweets
my_demo_list = []
with open('tweet_json_Barca.txt', encoding='utf-8') as json_file:  
    all_data = json.load(json_file)
    for each_dictionary in all_data:
        tweet_id = each_dictionary['id']
        text = each_dictionary['text']
        favorite_count = each_dictionary['favorite_count']
        retweet_count = each_dictionary['retweet_count']
        created_at = each_dictionary['created_at']
        my_demo_list.append({'tweet_id': str(tweet_id),
                             'text': str(text),
                             'favorite_count': int(favorite_count),
                             'retweet_count': int(retweet_count),
                             'created_at': created_at,
                            })
        #print(my_demo_list)
        tweet_json = pd.DataFrame(my_demo_list, columns = 
                                  ['tweet_id', 'text', 
                                   'favorite_count', 'retweet_count', 
                                   'created_at'])
```
